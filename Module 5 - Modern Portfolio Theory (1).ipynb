### Query 5.1 
5.1 For your chosen stock, calculate the mean daily return and daily standard deviation of returns, and then just annualise them to get mean expected annual return and volatility of that single stock. **( annual mean = daily mean * 252 , annual stdev = daily stdev * sqrt(252) )**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('PVR.csv')
dataset = dataset[dataset.Series == 'EQ']
dataset.dropna(inplace =True)
dataset = dataset['Close Price']
dataset
```

```python
ret = dataset.pct_change()
mean = ret.mean()
std = ret.std()
ann_mean = mean * 252
ann_std = std * np.sqrt(252)
print("Annual mean :",ann_mean)
print("Annual std :",ann_std)
```

```python
#Query2
data1 = pd.read_csv('LT.csv')
data2 = pd.read_csv('PNB.csv')
data3 = pd.read_csv('SUZLON.csv')
data4 = pd.read_csv('TCS.csv')
data5 = pd.read_csv('RAYMOND.csv')
data1 = data1[data1.Series == 'EQ']
data1.dropna(inplace =True)
data2 = data2[data2.Series == 'EQ']
data2.dropna(inplace =True)
data3 = data3[data3.Series == 'EQ']
data3.dropna(inplace =True)
data4 = data4[data4.Series == 'EQ']
data4.dropna(inplace =True)
data5 = data5[data5.Series == 'EQ']
data5.dropna(inplace =True)
datas = pd.DataFrame([data1['Close Price'], data2['Close Price'],data3['Close Price'],data4['Close Price'],data5['Close Price']]).transpose()
datas.columns = ['LT','PNB','SUZ','TCS','RAY']
```

```python
datas
```

```python
ret = datas.pct_change()
mean_daily = ret.mean()
cov = ret.cov()
weights = np.asarray([0.20,0.20,0.20,0.20,0.20])
portfolio_ret = np.sum(mean_daily*weights)*252
portfolio_std = np.sqrt(np.dot(weights.T,np.dot(cov,weights))) * np.sqrt(252)
print("Annualized mean return :",portfolio_ret)
print("Annualized Volatility :",portfolio_std)
```

```python
#Query 3 
from matplotlib.pyplot import figure
figure(figsize=(20,10))

num_portfolio = 25000
res = np.zeros((3+datas.columns.size,num_portfolio))
for i in range(num_portfolio):
    weights = np.array(np.random.random(datas.columns.size))
    weights/=np.sum(weights)
    portfolio_ret = np.sum(mean_daily*weights)*252
    portfolio_std = np.sqrt(np.dot(weights.T,np.dot(cov,weights))) * np.sqrt(252)
    res[0,i] = portfolio_ret
    res[1,i] = portfolio_std
    res[2,i] = res[0,i]/res[1,i]
    for j in range(len(weights)):
        res[j+3,i] = weights[j]
        
res_frame = pd.DataFrame(res.T,columns = ['ret','std','sharpe','LT','PNB','SUZ','TCS','RAY'])
max_sharpe = res_frame.iloc[res_frame['sharpe'].idxmax()]
min_vol = res_frame.iloc[res_frame['std'].idxmin()]
plt.scatter(res_frame['std'],res_frame.ret,c=res_frame.sharpe,cmap='RdYlBu')
plt.xlabel('Volatility')
plt.ylabel('Returns')
plt.colorbar()
#plotting red to highlight highest sharpe ratio
plt.scatter(max_sharpe[1],max_sharpe[0],marker=(5,1,0),color='r',s=200)
#plotting green to highlight position of minimum volatitility
plt.scatter(min_vol[1],min_vol[0],marker=(5,1,0),color='g',s=200)
```
