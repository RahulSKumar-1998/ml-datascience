```python
# Query 2
returns = datas.pct_change().mean()*252
returns = pd.DataFrame(returns)
returns.columns = ['Returns']
returns['Volatility'] = datas.pct_change().std() * np.sqrt(252)
returns
```

```python
# Query 3
data = np.asarray([np.asarray(returns['Returns']),np.asarray(returns['Volatility'])]).T
cdata = np.where(np.isnan(data),0,data)
cdata
```

```python
from sklearn.cluster import KMeans
x = cdata
wcss = []
for k in range(2,25):
    k_means = KMeans(n_clusters=k)
    k_means.fit(x)
    wcss.append(k_means.inertia_)

fig = plt.figure(figsize=(20,10))
plt.plot(range(2,25),wcss)
plt.grid(True)
plt.show()
```

```python
from scipy.cluster.vq import kmeans,vq
centroids,_ = kmeans(cdata,5)
idx,_ = vq(cdata,centroids)
data = cdata
```

```python
from matplotlib.pyplot import figure
figure(figsize = (20,10))
plt.scatter(x[idx==0,0],x[idx==0,1],s=100,c='blue',label='Cls1',alpha=0.50)
plt.scatter(x[idx==1,0],x[idx==1,1],s=100,c='red',label='Cls2',alpha=0.50)
plt.scatter(x[idx==2,0],x[idx==2,1],s=100,c='green',label='Cls3',alpha=0.50)
plt.scatter(x[idx==3,0],x[idx==3,1],s=100,c='cyan',label='Cls4',alpha=0.50)
plt.scatter(x[idx==4,0],x[idx==4,1],s=100,c='yellow',label='Cls5',alpha=0.50)
plt.scatter(centroids[:,0],centroids[:,1],s = 200,c='black',alpha = 0.8)
plt.show()
```
